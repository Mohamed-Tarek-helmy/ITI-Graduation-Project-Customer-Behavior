# Ass Brand 

```python
df_filter=pd.read_csv("datak.csv")
```

```python
df_filter
```

```python
df_filter_ass = df_filter.groupby(['user_id','brand']).sum()['purchase_count'].unstack().reset_index().fillna(0).set_index('user_id')
df_filter_ass
```

```python
# Encode
def encoder(x):
    if(x <= 0): return 0
    if(x >= 1): return 1
```

```python
# Apply to the dataframe
df_filter_ass_encoded = df_filter_ass.applymap(encoder)
df_filter_ass_encoded.head()
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itemsets = apriori(df_filter_ass_encoded,min_support=0.02, use_colnames=True)
frequent_itemsets
```

```python
# Create Association Rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Sort values based on lift
rules = rules.sort_values("lift",ascending=False).reset_index(drop= True)
rules
```
